
## 5. Tomando Decisiones: Condicionales (if, else)

Constantemente tomamos decisiones basadas en lo que observamos: *Si* empieza a llover, guardamos la libreta; *de lo contrario*, seguimos anotando. 

En Python, podemos enseñarle a la computadora a tomar decisiones lógicas similares usando **condicionales**. Utilizamos la palabra `if` (si ocurre esto...) y `else` (de lo contrario...). Esto es fundamental para filtrar datos, por ejemplo, separar las fotos de especies amenazadas del resto.

```python
# Supongamos que estamos analizando el peso de un animal capturado (en kg)
especie = "Jaguar"
peso_kg = 85

print("Analizando registro de:", especie)

# Le decimos a Python que tome una decisión basada en el peso
if peso_kg > 50:
    # Todo lo que esté "indentado" (con sangría) pertenece a esta decisión
    print("¡Atención! Es un animal de gran tamaño.")
   
else:
    # Si la condición de arriba NO se cumple, Python ejecuta esto:
    print("Es un animal mediano o pequeño.")
```

### ¿Y si tenemos más de dos opciones?
A veces las cosas no son blanco o negro. Si necesitamos evaluar múltiples condiciones, usamos `elif` (una abreviación de "else if", que significa "o si no, si...").

```python
# Evaluando el nivel de amenaza según la Lista Roja de la UICN
estado_conservacion = "VU" # VU = Vulnerable

if estado_conservacion == "CR":
    print("Peligro Crítico: Prioridad máxima de conservación.")
elif estado_conservacion == "EN":
    print("En Peligro: Monitoreo estricto requerido.")
elif estado_conservacion == "VU":
    print("Vulnerable: Especie con alto riesgo de pasar a peligro.")
else:
    print("Preocupación Menor o Datos Insuficientes.")
```

## 6. Funciones
Las funciones son bloques de código reutilizables. Piensa en ellas como una "receta": le das unos ingredientes (datos de entrada), hace un proceso y te devuelve un resultado (datos de salida). Las definimos con la palabra `def`.

```python
# Definimos una función para calcular densidad poblacional
def calcular_densidad(numero_individuos, area_km2):
    densidad = numero_individuos / area_km2
    return densidad

# Usamos la función con nuestros datos hipotéticos
jaguares_estimados = 12
area_estudio = 50 

resultado = calcular_densidad(jaguares_estimados, area_estudio)
print("La densidad es de", resultado, "jaguares por km cuadrado.")
```

## 6. Ejercicio Práctico: El Filtro de Detecciones
**Situación:** Tu cámara trampa generó una lista de todas las veces que el sensor de movimiento se activó. A veces fue un animal, pero a veces fue solo el viento moviendo la maleza (registrado como "Vacío").

**Tarea:** 1. Crea un ciclo `for` que revise cada elemento de la lista.
2. Usa un condicional `if` para verificar si la detección es "Jaguar".
3. Si es un Jaguar, suma 1 a un contador.

```python
# Lista de registros de la cámara trampa
registros = ["Vacío", "Puma", "Jaguar", "Vacío", "Tapir", "Jaguar", "Vacío", "Ocelote"]

# 1. Iniciamos nuestro contador en cero
contador_jaguares = 0

# 2. Creamos el ciclo
for registro in registros:
    # 3. Ponemos la condición
    if registro == "Jaguar":
        contador_jaguares = contador_jaguares + 1 # Sumamos 1 si se cumple

# 4. Mostramos el resultado final
print("El número total de fotos útiles de jaguares es:", contador_jaguares)
```

## 7. Visualizando Anotaciones (Bounding Boxes)

En visión por computador, le enseñamos a la máquina a extraer información útil mostrando ejemplos. Cuando queremos enseñarle a un modelo dónde está un animal, no le pasamos solo la foto; le pasamos la foto con una **anotación**. 

La forma más común de anotar es dibujando "cajas delimitadoras" (o *bounding boxes*) alrededor del animal. A diferencia de la segmentación de instancias, donde dibujaríamos la silueta o máscara exacta del individuo, una caja delimitadora es simplemente un rectángulo que enmarca a nuestro objeto de interés.

Vamos a usar una herramienta gráfica de Python llamada `matplotlib` para dibujar nuestra primera caja delimitadora.

### Visualizando Anotaciones de Datos Reales

En el mundo real, cuando descargamos un conjunto de datos (dataset) de cámaras trampa o drones, generalmente viene organizado en dos carpetas:
1. `images/`: Contiene las fotografías (ej. `frame_220920.jpg`).
2. `labels/`: Contiene archivos de texto con las coordenadas de las anotaciones (ej. `frame_220920.txt`).

**El formato de las etiquetas:**
El estándar más común (conocido como formato YOLO) guarda la información en el archivo de texto separada por espacios, en este orden:
`[clase] [centro_X] [centro_Y] [ancho] [alto]`

*Nota: Las coordenadas en este formato suelen estar "normalizadas" (van del 0 al 1, como si fueran porcentajes del tamaño de la imagen) para que el modelo funcione igual sin importar si la foto es gigante o pequeña.*

¡Vamos a programar a Python para que abra nuestra imagen, lea el archivo de texto y dibuje la caja delimitadora automáticamente!

```python
# Importamos las herramientas gráficas necesarias
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.image as mpimg # Nueva herramienta para leer imágenes
import os
```

```python
# 1. Definimos dónde están nuestros archivos reales
ruta_imagen = 'images/frame_220920.jpg'
ruta_etiqueta = 'labels/frame_220920.txt'
```

```python
# 2. Leemos la imagen
imagen = mpimg.imread(ruta_imagen)
alto_img, ancho_img, canales = imagen.shape # Obtenemos el tamaño real de la foto en píxeles

# 3. Leemos la etiqueta de texto
# Usamos 'with open' para abrir el archivo, leerlo y que se cierre automáticamente de forma segura
with open(ruta_etiqueta, 'r') as archivo:
    linea = archivo.readline().strip() # Leemos la primera línea y quitamos espacios extra
    
    # Separamos los valores. Recuerda que al leer un archivo de texto, todo es tipo "String" (texto)
    valores = linea.split()
    
    # Convertimos los textos a números (Enteros para la clase, Decimales/Floats para coordenadas)
    clase_animal = int(valores[0])
    centro_x_norm = float(valores[1])
    centro_y_norm = float(valores[2])
    ancho_norm = float(valores[3])
    alto_norm = float(valores[4])
```

```python
# 4. Matemáticas de campo: Convertir porcentajes (0 a 1) a píxeles reales
ancho_pixel = ancho_norm * ancho_img
alto_pixel = alto_norm * alto_img

# Para dibujar el rectángulo, Matplotlib necesita saber dónde empieza la esquina superior izquierda
x_min = (centro_x_norm * ancho_img) - (ancho_pixel / 2)
y_min = (centro_y_norm * alto_img) - (alto_pixel / 2)
```

```python
# 5. Visualizamos el resultado
fig, ax = plt.subplots(figsize=(12, 12))
ax.imshow(imagen) # Mostramos la foto real de fondo

# Creamos y añadimos la caja
caja = patches.Rectangle((x_min, y_min), ancho_pixel, alto_pixel, 
                         linewidth=2, edgecolor='blue', facecolor='none')
ax.add_patch(caja)

plt.axis('off') # Ocultamos los números de los ejes para verla como una foto normal
plt.show()
```
